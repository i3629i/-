```python
import requests
from bs4 import BeautifulSoup
import csv
import json

#융합보안학과 윤원석
page = requests.get('https://namu.wiki/w/크롤링')

soup = BeautifulSoup(page.content, "html.parser")

soup_test2 = soup.find_all('h2',attrs={'class':'wiki-heading'})
soup_test = soup.find_all('h2')

content = soup.find_all('div',attrs={'class':'wiki-heading-content'})

f = open('crawl_test99.csv','w',newline='',encoding="UTF-8")
wr = csv.writer(f)
for soup_test,content in zip(soup_test[1:],content):
    wr.writerow([soup_test.text.split('[편집]')[0],content.text])
f.close()


```

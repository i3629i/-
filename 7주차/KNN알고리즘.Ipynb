# KNN(K-Nearest Neighbor) 알고리즘

## 1. KNN개념   
* <span style="color:Yellow">지도학습으로 데이터가 주어지면 그 주변 데이터를 살펴본뒤 더 많은 데이터가 포함되어 있는 범주로 분류</span>
* K개의 데이터를 본뒤 주변 데이터에 포함되어있는 클래스가 많은 것으로 클래스 분류를 진행함
* KNN은 별도의 train학습이 필요없기 때문에 SVM이나 선형회기보다 빠름

#### 1.3.1 유클리드 거리 알고리즘(연속형)
* N과 데이터간의 거리를 직선으로 측정 / X값의 증가량, Y값의 증가량 제곱해서 루트
$$f(x) = \sqrt{(Ax_1-Bx_2)^2 + (Ay_1-By_2)^2}$$      

* 3차원에서 유클리드 거리 알고리즘  
$$f(x) = \sqrt{(Ax_1-Bx_2)^2 + (Ay_1-By_2)^2 + (Ay_3-By_3)^2}$$      

![image.png](./img/euclid.png)
* A ~ N 3.16 / B ~ N : 2.83

### 1.4 Data 정규화
* 데이터 값의 스케일이 심하게 차이나는 경우에는 문제가 생길 수 있기 때문에 정규화가 필요함

#### 1.4.1 Min_Max
* 0과 1 사이의 값으로 변환   
$$f(x) = \frac{(x-min)}{(max-min)}$$

*  100개중 99개가 X나 Y의 값이 0~40 사이인데 하나가 100을 차지해버리면, MAX값 하나를 사용하기 땜누에 편향 되어버림.

#### 1.4.2 Z-Score(Z점수 정규화)
* Min_Max의 이상치 문제를 피하기위한 전략
* Z값은 양수나 음수 모두 가능한데, 음수이면 표본 값이 평균보다 나다는 뜻이고, 양수이면 표본 값이 평균보다 높다는 뜻


$$f(x) = \frac{(x-평균)}{(표준편차)}$$

# Kmeans 알고리즘

## 1. K means
* 비지도학습 알고리즘으로 주어진 데이터로부터 그룹의 갯수를 지정하고, 클러스터의 중심과 데이터들간의 평균거리를 구함

<img src="./img/WL1tIZ4.gif.gif" width="60%" height="60%">   
   
### 1.1 K means의 순서
(1) 데이터셋에서 K개의 그룹을 지정한다   
(2)  Centroids 각 데이터들과 가장 가까운 그룹에 속한 그룹에 할당한다 // 예를들어 데이터를 넣었는데 가장 가까운 그룹 클래스가 '과일'  
(3) 2번 과정에서의 결과를 바탕으로 랜덤으로 Centroids를 새롭게 지정    
(4) 가장 작은 Centroids를 선택

### 1.2 K means 알고리즘 원리
Input :    K, 데이터   

* 식 알고리즘 순서

$$ J() = \frac{1}{m}  \sum\limits_{i=1}^{m}{||x^(i) - u_c(i)||^2}$$   

<img src="./img/fomula3.jpg" width="50%" height="50%">   


<span style="color:yellow; font-size: large;">① 2차원에서 rnk : n번째 데이터가 k번째에 cluster에 속하면 1, 아니면 0 | 즉, 가장가까운 클러스터 선택</span>   
<span style="color:yellow; font-size: large;">② rnk값(1혹은0)과 데이터의 x,y값을 곱해 평균을 구한값을 다음 Centroid1(x,y), Centroid2(x,y) 의 좌표값으로 지정 </span>   
<span style="color:yellow; font-size: large;">계속 반복해  Centroid1(x,y), Centroid2(x,y)가 더이상 낮아지지 않을 때까지 반복 </span>   

* 예제

데이터 예제   
   
|Data|X|Y|
|---|---|---|
|x1|1|1|
|x2|6|6|
|x3|7|7|
|x4|3|2|
|x5|6|8|
|x6|2|1|
|x7|2|3|   

(1) Centroid1 = (4,3), Centroid2 = (7,9) 랜덤으로 지정   
(2) 유클리드 거리를 이용해 Centroid1(4,3)와 Data(x,y) Centroid2(7,9)와 Data(7,9)의 점과점 사이의 거리 모두 계산
   
|Data|s1(그룹1)|s2(그룹2)|
|---|---|---|
|x1|3.6056|10|
|x2|3.6056|3.1623|
|x3|5|2|
|x4|1.4142|8.0623|
|x5|5.3852|1.4142|
|x6|2.8284|9.4340|
|x7|2|7.8102|   
   
(3) 위의 <span style="color:yellow; font-size: large;">②</span> 번 식을 통해서 그룹1에 더 가까운 값의 X데이터 끼리 더해 평균을 구함   
* s1그룹의 경우 x1, x4, x6, x7이 그룹2보다 작아 군집됨
    * x1(1,1) + x4(3,2) + x6(2,1) + x7(2,3)의 평균 = (2, 1.75) <<==이게 새로운 Centroid위치가 되는 것
   
       
* 반대로 s2는 x2, x3, x5가 그룹2에 군집
    * 새로운 Centroid2 == (6.3333,7)
     
(4) 위의 상황을 반복해 Centroid1, Centroid2 의 좌표값이 변하지 않을 때 알고리즘을 종료한다.

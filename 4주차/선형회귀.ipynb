```python
# MSE(평군제곱오차)를 이용해 직선의 방정식의 과의 오차 평균을 구함
# N개의 데이터, 직선의 방정식 - 13x + 1,
# ∑(해당 직선의 값 - 실제 값)^2 / N

import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

t_x = np.arange(0,7)
t_y = 13 * t_x + 1

plt.plot(t_x,t_y)
plt.plot([2,3,4,5],[25,50,42,61],'ro')
```

```python
df = pd.read_csv("hw_200.csv")

# 해당 컬럼 삭제
df = df.drop(["Index"], axis = 1)

df.sample(5)

X = df["Height(inches)"]
Y = df["Weight(pounds)"]
plt.plot(X,Y, 'go')


line_filtter = LinearRegression()
line_filtter.fit(X.values.reshape(-1,1), Y)

# 원하는 X값을 넣어 Y(무게를)를 예측 할 수 있다. 70인치의 키를 가진 사람의 몸무게예측
print(line_filtter.predict([[70]]))
```

```python
#Kg, Cm로 변환해서 선형변환

Cm_X = X * 2.54
Kg_Y = Y * 0.453592 

line_filtter = LinearRegression()
line_filtter.fit(Cm_X.values.reshape(-1,1), Kg_Y)

plt.plot(Cm_X, Kg_Y, 'go')
plt.plot(Cm_X,line_filtter.predict(Cm_X.values.reshape(-1,1)))
plt.show()

print(line_filtter.predict([[178]]), "kg")
```

```python
df = pd.read_csv("salary_data.csv")
df.sample(10)

sal_X = df["YearsExperience"]
sal_Y = df["Salary"]

line_filtter = LinearRegression()
line_filtter.fit(sal_X.values.reshape(-1,1), sal_Y)

plt.plot(sal_X,sal_Y,'o')
plt.plot(sal_X,line_filtter.predict(sal_X.values.reshape(-1,1)))
```

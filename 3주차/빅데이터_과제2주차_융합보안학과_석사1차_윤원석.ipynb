```python
from selenium import webdriver
import time
import json
import csv

start_page = 1
end_page = 10

f = open('crawl_dataset.csv', 'w', newline='', encoding='UTF-8')
wr = csv.writer(f)

for current_page in range(start_page, end_page+1):
    url = "https://data.gg.go.kr/portal/data/dataset/searchDataset.do?page=" + str(current_page)
    
    options = webdriver.ChromeOptions()
    options.add_argument('window-size=1920,1080')
    
    driver = webdriver.Chrome('chromedriver', options=options)
    driver.get(url)
    driver.set_page_load_timeout(10)
    table = driver.find_element_by_xpath('/html/body') #xpath를 이용해서 가져옵니다.
   
    jsons = json.loads(table.text)
    jsons = jsons['data']
    wr.writerow("")
    for k in jsons:
        wr.writerow([k['infNm'],k['infExp']])
    driver.quit()
    print(current_page, "페이지 크롤 완료")
    
f.close() 
```

```python
from selenium import webdriver
import time
import json
import csv

print("키워드 목록 ")
print("전체, 건강, 경제, 교육, 규제개혁, 복지, 안전, 여가, 보육, 일자리, 주택, 행정재정, 환경")
keyword = input('카테고리 입력 = ')

def switch(x):
    return {'전체':'00000','건강':'00026','경제':'00034','교육':'00012','규제개혁':'00027','복지':'00016','안전':'00014','여가':'00009','보육':'00031','일자리':'00025','주택':'00010','행정재정':'00035','환경':'00086'}[x]
    
print(keyword)
print(switch(keyword))    

f = open('crawl_dataset.csv', 'w', newline='', encoding='UTF-8')    
wr = csv.writer(f)

start_page = 1
end_page = 10

#url = 'https://www.open.go.kr/com/tema/searchTopkwd.ajax?viewPage=3'
# url = 'https://www.open.go.kr/com/tema/temaSrhList.do?topictype='+switch(keyword)+'#1'

for current_page in range(start_page, end_page+1):
    url = "https://www.open.go.kr/com/tema/searchTopkwd.ajax?topictype="+switch(keyword)+"&site=nflst&kwd=&insttSeCd=&eduYn=N&startDate=20210222&endDate=20210323&insttCdNm=&insttCd=&searchMainYn=&searchCurrTab=1&viewPage="+str(current_page)+"&rowPage=10&sort=s&searchAll="
    options = webdriver.ChromeOptions()
    options.add_argument('window-size=1920,1080')
    driver = webdriver.Chrome('chromedriver', options=options)
    
    driver.set_page_load_timeout(10)
    
    driver.get(url)
    
    table = driver.find_element_by_xpath('/html/body') #xpath를 이용해서 가져옵니다.
   
    jsons = json.loads(table.text)
    jsons = jsons['rtnList']
    wr.writerow("")
    for k in jsons:
#         print(k)
        wr.writerow([k['ALL_PROC_INSTT_NM'],k['INFO_SJ']])
    driver.quit()
    print(current_page, "페이지 크롤 완료")
    
f.close() 
```
